````python
# Complete notebook cell: Enrich Bhagavad Gita CSV with Sarvam API structured JSON output

import sys
import subprocess
import importlib
import importlib.util

# First, ensure pydantic and pydantic-core are compatible
print("Resolving pydantic dependencies...")
subprocess.check_call([
    sys.executable, "-m", "pip", "install", 
    "--upgrade", "--force-reinstall",
    "pydantic>=2.0",
    "pydantic-core>=2.23.0"
])

required_packages = {
    "pandas": "pandas",
    "tqdm": "tqdm",
    "dotenv": "python-dotenv",
    "requests": "requests",
    "pydantic": "pydantic",
}

for import_name, package_name in required_packages.items():
    if importlib.util.find_spec(import_name) is None:
        print(f"Installing missing package: {package_name}")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "--upgrade", package_name])

# Clear cached modules
importlib.invalidate_caches()
for module_name in list(sys.modules):
    if (
        module_name == "google"
        or module_name.startswith("google.")
        or module_name == "pydantic"
        or module_name.startswith("pydantic.")
        or module_name == "pydantic_core"
        or module_name.startswith("pydantic_core.")
    ):
        sys.modules.pop(module_name, None)

import os
import re
import json
import time
from pathlib import Path
from typing import List

import pandas as pd
from tqdm.auto import tqdm
from dotenv import load_dotenv
from pydantic import BaseModel, Field
import requests

load_dotenv()

SARVAM_API_KEY = os.getenv("SARVAM_API_KEY")
if not SARVAM_API_KEY:
    raise ValueError("SARVAM_API_KEY not found. Add it to your .env file or environment variables.")

MODEL_NAME = "sarvam-m"
TEMPERATURE = 0.4
SARVAM_API_URL = "https://api.sarvam.ai/v1/chat/completions"

INPUT_CSV = Path("../Bhagwad_Gita.csv")
OUTPUT_CSV = Path("../enriched_gita.csv")
AUTOSAVE_EVERY = 20
MAX_RETRIES = 3
RETRY_BACKOFF_SECONDS = 3

REQUIRED_INPUT_COLUMNS = ["ID", "Chapter", "Verse", "Shloka", "Transliteration", "HinMeaning", "EngMeaning", "WordMeaning"]
NEW_COLUMNS = ["Interpretation", "Topics", "EmotionTags", "Summary"]

PROMPT_TEMPLATE = """
You are an expert Bhagavad Gita philosophical interpreter and spiritual knowledge assistant.

Your task is to deeply analyze a Bhagavad Gita verse and generate a modern philosophical interpretation suitable for:
- AI-powered semantic retrieval
- spiritual guidance systems
- emotionally intelligent RAG applications
- educational and reflective use

INSTRUCTIONS:
1. Carefully understand the verse, translation, and word meanings.
2. Infer the deeper teaching being communicated.
3. Explain the meaning in modern, clear, natural English.

OUTPUT REQUIREMENTS:
Return ONLY valid JSON.

JSON schema:
{{
  "interpretation": "...",
  "topics": ["...", "..."],
  "emotion_tags": ["...", "..."],
  "summary": "..."
}}

FIELD REQUIREMENTS:
- interpretation: 50-80 words, modern English, practical and philosophical
- topics: 3-5 items, lowercase, concise philosophical themes
- emotion_tags: 2-4 items, lowercase, emotional or psychological states
- summary: one sentence, under 20 words, concise teaching

Verse: {verse}
English Meaning: {translation}
Word Meaning: {word_meaning}
"""

class GitaEnrichment(BaseModel):
    interpretation: str = Field(..., min_length=20)
    topics: List[str] = Field(..., min_length=3, max_length=5)
    emotion_tags: List[str] = Field(..., min_length=2, max_length=4)
    summary: str = Field(..., min_length=10)

def clean_json_response(text: str) -> str:
    """Remove markdown fences and return clean JSON."""
    if not isinstance(text, str):
        raise ValueError("Sarvam response is not a string.")
    cleaned = text.strip()
    cleaned = re.sub(r"^```(?:json)?\s*", "", cleaned, flags=re.IGNORECASE)
    cleaned = re.sub(r"\s*```$", "", cleaned)
    start = cleaned.find("{")
    end = cleaned.rfind("}")
    if start == -1 or end == -1:
        raise ValueError("No valid JSON object found.")
    return cleaned[start:end + 1]

def normalize_list(values: List[str]) -> List[str]:
    """Normalize topic/emotion tag lists."""
    cleaned = []
    seen = set()
    for value in values:
        item = str(value).strip().lower()
        if item and item not in seen:
            cleaned.append(item)
            seen.add(item)
    return cleaned

def validate_enrichment(raw_data: dict) -> GitaEnrichment:
    """Validate Sarvam response."""
    parsed = GitaEnrichment.model_validate(raw_data)
    parsed.topics = normalize_list(parsed.topics)
    parsed.emotion_tags = normalize_list(parsed.emotion_tags)
    return parsed

def generate_enrichment(verse: str, translation: str, word_meaning: str) -> GitaEnrichment:
    """Generate interpretation using Sarvam API."""
    prompt = PROMPT_TEMPLATE.format(
        verse=str(verse or "").strip(),
        translation=str(translation or "").strip(),
        word_meaning=str(word_meaning or "").strip(),
    )
    last_error = None
    for attempt in range(1, MAX_RETRIES + 1):
        try:
            headers = {
                "Authorization": f"Bearer {SARVAM_API_KEY}",
                "Content-Type": "application/json"
            }
            payload = {
                "model": MODEL_NAME,
                "messages": [
                    {"role": "user", "content": prompt}
                ],
                "temperature": TEMPERATURE,
                "max_tokens": 1000
            }
            response = requests.post(SARVAM_API_URL, headers=headers, json=payload, timeout=30)
            response.raise_for_status()
            result = response.json()
            
            response_text = result["choices"][0]["message"]["content"]
            json_text = clean_json_response(response_text)
            raw_data = json.loads(json_text)
            validated = validate_enrichment(raw_data)
            return validated
        except Exception as exc:
            last_error = exc
            print(f"Attempt {attempt}/{MAX_RETRIES} failed: {exc}")
            if attempt < MAX_RETRIES:
                sleep_for = RETRY_BACKOFF_SECONDS * attempt
                print(f"Retrying in {sleep_for} seconds...")
                time.sleep(sleep_for)
    raise RuntimeError(f"Failed after {MAX_RETRIES} retries: {last_error}")

def row_already_processed(row: pd.Series) -> bool:
    """Check whether row already contains generated values."""
    return all(
        col in row.index and pd.notna(row[col]) and str(row[col]).strip()
        for col in NEW_COLUMNS
    )

def save_dataframe(df: pd.DataFrame, path: Path):
    """UTF-8 safe saving."""
    df.to_csv(path, index=False, encoding="utf-8-sig")

print("Running example test...")
example_result = generate_enrichment(
    verse="कर्मण्येवाधिकारस्ते मा फलेषु कदाचन ।",
    translation="You have a right to perform your prescribed duties, but you are not entitled to the fruits of your actions.",
    word_meaning="karmaṇi = prescribed duties"
)
print("\nExample Output:")
print(json.dumps(example_result.model_dump(), ensure_ascii=False, indent=2))

if not INPUT_CSV.exists():
    raise FileNotFoundError(f"Input file not found: {INPUT_CSV.resolve()}")

source_csv = OUTPUT_CSV if OUTPUT_CSV.exists() else INPUT_CSV
print(f"\nLoading dataset from: {source_csv.resolve()}")
df = pd.read_csv(source_csv, encoding="utf-8-sig")

missing_columns = [col for col in REQUIRED_INPUT_COLUMNS if col not in df.columns]
if missing_columns:
    raise ValueError(f"Missing required columns: {missing_columns}")

for col in NEW_COLUMNS:
    if col not in df.columns:
        df[col] = ""

print(f"Rows loaded: {len(df)}")
processed_count = sum(row_already_processed(row) for _, row in df.iterrows())
print(f"Rows already processed: {processed_count}")

failed_rows = []
processed_since_save = 0

for idx, row in tqdm(df.iterrows(), total=len(df), desc="Enriching verses"):
    if row_already_processed(row):
        continue
    try:
        result = generate_enrichment(
            verse=row["Shloka"],
            translation=row["EngMeaning"],
            word_meaning=row["WordMeaning"],
        )
        df.at[idx, "Interpretation"] = result.interpretation
        df.at[idx, "Topics"] = json.dumps(result.topics, ensure_ascii=False)
        df.at[idx, "EmotionTags"] = json.dumps(result.emotion_tags, ensure_ascii=False)
        df.at[idx, "Summary"] = result.summary
        processed_since_save += 1
        print(f"Processed row {idx} | Chapter {row['Chapter']} Verse {row['Verse']}")
        if processed_since_save >= AUTOSAVE_EVERY:
            save_dataframe(df, OUTPUT_CSV)
            print(f"Autosaved after {processed_since_save} rows.")
            processed_since_save = 0
    except Exception as exc:
        failed_rows.append({
            "index": idx,
            "ID": row.get("ID"),
            "Chapter": row.get("Chapter"),
            "Verse": row.get("Verse"),
            "error": str(exc),
        })
        print(f"Failed row {idx} | Chapter {row.get('Chapter')} Verse {row.get('Verse')}: {exc}")

save_dataframe(df, OUTPUT_CSV)
print("\nEnrichment completed.")
print(f"Final output saved to: {OUTPUT_CSV.resolve()}")

if failed_rows:
    print("\nFailed rows:")
    for failure in failed_rows:
        print(failure)
else:
    print("\nNo failed rows.")

generated_rows = df[
    df["Interpretation"].notna()
    & df["Interpretation"].astype(str).str.strip().astype(bool)
].tail(3)

print("\nSample Generated Output:")
display(generated_rows[["Chapter", "Verse", "Shloka", "Interpretation", "Topics", "EmotionTags", "Summary"]])
````
